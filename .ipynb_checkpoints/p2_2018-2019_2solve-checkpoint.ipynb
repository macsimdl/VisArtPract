c) Can we convolve a color (RGB) image? In order to convolve an RGB each of the channels must be convolved separately.

Implement a function that applies a 3-channel filter and visualize its result on `face.png` with a 15x15 uniform mask.

Which is the result obtained when applying a single-channel filter channel by channel separately?

We get a blur of the image in every chanel, and then combined it results in a blur of the rgb image. 

```python
def convolve_3chanels(image):
    #Mask
    mask_h1d2 =np.ones((15,15))
    #normalize 2d mask
    mask_h1d2 /= mask_h1d2.sum() 
    #Convolution channel by channel
    image = image.copy()
    image[:,:,0] = nd.convolve(image[:,:,0],mask_h1d2)
    image[:,:,1] = nd.convolve(image[:,:,1],mask_h1d2)
    image[:,:,2] = nd.convolve(image[:,:,2],mask_h1d2)

    return image
#
con_face3 = convolve_3chanels(face)
plt.imshow(con_face3,  vmin = 0, vmax=1)
plt.show()

#Implementation of a 3-channel filter.
#mask_h1d=np.ones((15,15,3))

fig,(ax,ax1) = plt.subplots(nrows=1,ncols=2,figsize=(15,2))

#Original Image
ax.axis('off')
ax.imshow(face, vmin = 0, vmax=255)
ax.set_title('Original Image')

#Convolution channel by channel Image
ax1.axis('off')
ax1.imshow(con_face3, vmin = 0, vmax=255)
ax1.set_title('Convolution channel by channel')

```

### 2.6 Applying smoothing in order to obtain hybrid images

a) Given `Einstein.jpg` and `monroe.jpg` images:

1. Plot both images.
2. Apply a low-pass filter to both of them and plot them.
3. Apply a high-pass filter to both of them and plot them.

By applying smoothing over an image I, we apply a `low-pass` filter. The resulting image can be called L(I). If we substract the filtered one from the original image, we obtain its high frequencies, that we can call H(I), i.e. we apply a `high-pass` filter. 

***H(I) = I - L(I)***

Hint: In order to highlight the effect, in the L(I1) image you should define a lower sigma, while for the H(I1) you should use a higher one. 

```python
einstein = io.imread("images/Einstein.jpg", cmap="gray")
monroe = io.imread("images/monroe.jpg", cmap="gray")

fig,(ax,ax1) = plt.subplots(nrows=1,ncols=2,figsize=(10,10))

ax.axis('off')
ax.imshow(einstein, cmap='gray')
ax.set_title("EINSTEIN")

ax1.axis('off')
ax1.imshow(monroe, cmap='gray')
ax1.set_title("MOROE")

#Images shown
plt.show()

einstein_low = fl.gaussian(einstein, 3)
monroe_low = fl.gaussian(monroe, 3)

fig,(ax,ax1) = plt.subplots(nrows=1,ncols=2,figsize=(10,10))

ax.axis('off')
ax.imshow(einstein_low, cmap='gray')
ax.set_title("EINSTEIN")

ax1.axis('off')
ax1.imshow(monroe_low, cmap='gray')
ax1.set_title("MOROE")

einstein_high = fl.laplace(einstein, 10)
monroe_high = fl.laplace(monroe, 10)

fig,(ax,ax1) = plt.subplots(nrows=1,ncols=2,figsize=(10,10))

ax.axis('off')
ax.imshow(einstein_high, cmap='gray')
ax.set_title("EINSTEIN")

ax1.axis('off')
ax1.imshow(monroe_high, cmap='gray')
ax1.set_title("MOROE")

#Images shown
plt.show()
```
